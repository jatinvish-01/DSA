## Arrays in Python

1. Using Array module 

```python
import array
```

```python
arr = array.array('i', [10, 20, 30, 40, 50])
print(arr)
print(type(arr))
```

```python
arr2 = array.array('i', [60,70,80])
print(arr2)
```

```python
# to access
arr[2]
```

```python
# traversing 
for i in arr:
      print(i, end=" ")
```

```python
# insertion 
arr.insert(3, 35)
print(arr)
```

```python
# deletion
arr.remove(30)
print(arr)
```

```python
# updation 
arr[2] = 30
print(arr)
```

```python
# sorting 

arr3 = array.array('i', [5,3,8,5,9,1])
print(arr3)
```

```python
sorted(arr3) #- it will give list 

# for array
sorted_arr3 = array.array('i', sorted(arr3))
print(sorted_arr3)
```

```python
# merging
merge = arr + arr2
```

```python
# splitting 
ix = int(len(merge) / 2)
```

```python
merge[:ix]
```

```python
merge[ix:]
```

2. Using Numpy

```python
import numpy as np
```

```python
np1 = np.array([10,20,30,40,50])
np2 = np.array([60,70,80])    
```

```python
np1[3]
```

```python
for i in np1:
      print(i, end=" ")
```

```python
np.insert(np1, 3, 35)
```

```python
np.delete(np1, 3)
```

```python
np1[2] = 35
print(np1)
```

```python
np3 = np.array([4,9,6,2,7,5])
np3
```

```python
np.sort(np3)
```

```python
merge = np.concatenate((np1, np2))
merge
```

```python
split1, split2 = np.array_split(merge, 2)      
```
