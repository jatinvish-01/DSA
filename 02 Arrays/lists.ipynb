### 2D Lists

```python
import numpy as np
```

```python
matrix = [
      [1, 2, 3],
      [4, 5, 6],
      [7, 8, 9]
]

print(type(matrix))
```

```python
# display
matrix
```

```python
# iteration
for i in matrix:
      print(i)
```

```python
# display with function

def display_matrix(mat):
      for i in mat:
            print(i)
```

```python
display_matrix(matrix)
```

```python
# traversal - to access elements of rows

for row in matrix:
      for ele in row:
            print(ele, end=' ')
      print()
```

```python
# using by indexes

for i in range(len(matrix)):
      print(matrix[i])
```

```python
m = len(matrix) #-> row
n = len(matrix[0]) #-> col

for i in range(m):
      for j in range(n):
            print(matrix[i][j], end=' ')
      print()
```

```python
print(matrix[2][1])
```

```python
matrix[1][2] = 0
display_matrix(matrix)
```

```python
# insertion

matrix[2].insert(1, 15)
display_matrix(matrix)
```

```python
# append
matrix[0].append(20)
display_matrix(matrix)
```

```python
matrix[2].remove(15)
display_matrix(matrix)
```

```python
matrix[0].pop()
display_matrix(matrix)
```
