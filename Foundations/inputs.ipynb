```python
# Input

numbers = input()
print(numbers[0])
print(type(numbers))
```

```python
# split - give the list of each elements

number = input().split()
print(numbers)
print(type(numbers))
```

```python
# map

number = list(map(int, input().split()))
print(number)
print(type(number))
```

```python
# some operationss

num = sum(map(int, input().split()))
print(num)
print(type(num))
```

```python
# que 01 - we have 2 dice and rolls it for 5 times
# 5
# 4 5 
# 3 3
# 6 1 
# 2 4
# 5 1

n = int(input())

for i in range(n):
      total = sum(map(int, input().split()))
      if total > 8:
            print(f"Roll {i+1}: Success")
      else:
            print(f"Roll {i+1}: Failed")
```
